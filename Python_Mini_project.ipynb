```python
#### Checking  Nulls values in columns
def null(df):
    null = df.isna().sum()
    total = len(df)
    percentage = null/total
    percentage = percentage *100
    data = pd.DataFrame({'null(numbers)': null, 'percentage(%)': percentage.round(2) })
    return data
null(df)
```

```python
#removing null data:
df=df.dropna()
for c in df.columns: 
  if df[c].dtype == np.float64: 
    df[c] = df[c].astype(int)
df
```

```python
import plotly.express as px
def top_10_illnesses(country):
    df_country = df[df['Country/Territory'] == country]
    min_year, max_year = df_country['Year'].min(), df_country['Year'].max()
    fig = px.bar(df_country.iloc[:, 2:].sum().sort_values(ascending=False).head(10), title='Top 10 Cause of Deaths in {} between {} and {}'.format(country, min_year, max_year), color_discrete_sequence=px.colors.qualitative.Pastel, labels={'value': 'Amount of deaths', 'index': 'Cause of Deaths'})
    
    fig.show()
```

```python
top_10_illnesses('India')
```

```python
#finding the relative increase in diseases over the years
df3=df[['Year','Cardiovascular Diseases','Neoplasms']]
df3=df3.groupby('Year').sum()
df3=df3.reset_index()
difference = df3.pct_change(axis=0);
difference=difference.drop(['Year'], axis=1)
temp=df3[['Year']]
df3=temp.join(difference)
df3
```
